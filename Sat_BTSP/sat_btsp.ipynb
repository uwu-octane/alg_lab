```python
from typing import Set
import random
import time

from sat_btsp import draw_edges, Node, BTSPSolverSat, Greedy_Btsp, ChristofidesAlg

from DBST.dbst_sat import  GreedyDBST
```

# Solving BTSP using (Cardinality-)SAT

## Point/Coordinate Generators

```python
def random_points(n, w=10_000, h=10_000) -> Set[Node]:
    """
    n random points with integer coordinates within the w * h rectangle.
    :param n: Number of points
    :param w: The width of the rectangle.
    :param h: The height of the rectangle.
    :return: A set of points as (x,y)-tuples.
    """
    return set((random.randint(0,w), random.randint(0,h)) for _ in range(n))
```

## DBST Solver

```python
#loop for 10 times and get the average time, put the result in a list
j = 10
average_result = {}
while j != 0:
    for i in range(20, 30):
        print("start loop", j, " for points: ", i)
        points = random_points(i)
        solver = BTSPSolverSat(points, "Gluecard4")

        start_time = time.time()
        result = solver.solve()
        end_time = time.time()

        run_time = end_time - start_time
        if j == 1 and i == 149:
            draw_edges(result)
        if i in average_result:
            average_result[i].append(run_time)
        else:
            average_result[i] = [run_time]
        #print(f"Time for points: {i, run_time}")
        #print("---------------------------------------------")
    j -= 1

# 计算平均值
for num_points, times in average_result.items():
    average_time = sum(times) / len(times)
    print(f"Average time for {num_points} points: {average_time}")
    print("---------------------------------------------")

   # if result:
    #    draw_edges(result)
```

## Greedy Solver

```python
j = 10
average_result = {}
while j != 0:
    for i in range(20, 30):
        print("start loop", j, " for points: ", i)
        points = random_points(i)
        greddy_btsp = Greedy_Btsp(points)
        edges = greddy_btsp.solve()
        solver = BTSPSolverSat(points, "Gluecard4",edges)

        start_time = time.time()
        #draw_edges(edges)
        result = solver.solve()
        end_time = time.time()

        run_time = end_time - start_time
        if j == 1 and i == 149:
            draw_edges(result)
        if i in average_result:
            average_result[i].append(run_time)
        else:
            average_result[i] = [run_time]

        #print(f"Time for points: {i, run_time}")
        #print("---------------------------------------------")
    j -= 1

# 计算平均值
for num_points, times in average_result.items():
    average_time = sum(times) / len(times)
    print(f"Average time for {num_points} points: {average_time}")
    print("---------------------------------------------")
```

## Greedy Solver with lb

```python
j = 10
average_result = {}
while j != 0:
    for i in range(3, 30):
        print("start loop", j, " for points: ", i)
        points = random_points(i)
        #usere greedy upper bound
        greedy_btsp = Greedy_Btsp(points)
        #gegebene greedy für DBST lower bound
        greedy_dbst = GreedyDBST(points,len(points))
        lb = greedy_dbst.solve()
        edges = greedy_btsp.solve()
        solver = BTSPSolverSat(points, "Gluecard4")

        start_time = time.time()
        #draw_edges(edges)
        result = solver.solve(lb,edges)

        end_time = time.time()
        run_time = end_time - start_time
        if j == 1 and i == 19:
            draw_edges(result)
        if i in average_result:
            average_result[i].append(run_time)
        else:
            average_result[i] = [run_time]

        #print(f"Time for points: {i, run_time}")
        #print("---------------------------------------------")
    j -= 1

# 计算平均值
for num_points, times in average_result.items():
    average_time = sum(times) / len(times)
    print(f"Average time for {num_points} points: {average_time}")
    print("---------------------------------------------")
```
